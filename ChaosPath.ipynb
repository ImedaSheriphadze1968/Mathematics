```python
import math
import matplotlib.pyplot as plt

def collatz_sequence(n):
    """–í—ã—á–∏—Å–ª—è–µ—Ç –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å –ö–æ–ª–ª–∞—Ç—Ü–∞ –¥–ª—è –∑–∞–¥–∞–Ω–Ω–æ–≥–æ —á–∏—Å–ª–∞."""
    seq = [n]
    while n != 1:
        n = n // 2 if n % 2 == 0 else 3 * n + 1
        seq.append(n)
    return seq

def stability_index(seq):
    """–í—ã—á–∏—Å–ª—è–µ—Ç –∏–Ω–¥–µ–∫—Å —Å—Ç–∞–±–∏–ª—å–Ω–æ—Å—Ç–∏: S(n) = max / –¥–ª–∏–Ω–∞"""
    return max(seq) / len(seq)

def analyze_and_plot(n):
    """–ê–Ω–∞–ª–∏–∑–∏—Ä—É–µ—Ç –∏ –≤–∏–∑—É–∞–ª–∏–∑–∏—Ä—É–µ—Ç –ö–æ–ª–ª–∞—Ç—Ü-–ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å."""
    seq = collatz_sequence(n)
    s_index = stability_index(seq)

    print(f"\nüîç –ê–Ω–∞–ª–∏–∑ –¥–ª—è —á–∏—Å–ª–∞: {n}")
    print(f"üìè –î–ª–∏–Ω–∞ –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç–∏: {len(seq)}")
    print(f"üìà –ú–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ: {max(seq)}")
    print(f"üßÆ –ò–Ω–¥–µ–∫—Å —Å—Ç–∞–±–∏–ª—å–Ω–æ—Å—Ç–∏ S(n): {round(s_index, 4)}")

    plt.figure(figsize=(10, 5))
    plt.plot(seq, marker='o', color='blue', label='–ó–Ω–∞—á–µ–Ω–∏—è –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç–∏')
    plt.title(f'–ö–æ–ª–ª–∞—Ç—Ü-–ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å –¥–ª—è n = {n}')
    plt.xlabel('–®–∞–≥')
    plt.ylabel('–ó–Ω–∞—á–µ–Ω–∏–µ')
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()

def batch_stability_plot(max_n=1000, A=1.5):
    """–°—Ç—Ä–æ–∏—Ç –≥—Ä–∞—Ñ–∏–∫ –∏–Ω–¥–µ–∫—Å–æ–≤ —Å—Ç–∞–±–∏–ª—å–Ω–æ—Å—Ç–∏ S(n) –¥–ª—è –¥–∏–∞–ø–∞–∑–æ–Ω–∞ —á–∏—Å–µ–ª."""
    x_vals, s_vals, bounds = [], [], []
    for i in range(2, max_n + 1):
        seq = collatz_sequence(i)
        x_vals.append(i)
        s_vals.append(stability_index(seq))
        bounds.append(A * math.log(i))

    plt.figure(figsize=(12, 6))
    plt.scatter(x_vals, s_vals, s=1, label='S(n)', color='darkblue')
    plt.plot(x_vals, bounds, color='red', linewidth=1, label=f'{A}¬∑log(n)')
    plt.title(f'üìâ –ò–Ω–¥–µ–∫—Å —Å—Ç–∞–±–∏–ª—å–Ω–æ—Å—Ç–∏ S(n) –¥–æ {max_n}')
    plt.xlabel('n')
    plt.ylabel('S(n)')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

def main():
    print("üî¢ –î–æ–±—Ä–æ –ø–æ–∂–∞–ª–æ–≤–∞—Ç—å –≤ Collatz Insight")
    print("–í–≤–µ–¥–∏—Ç–µ —á–∏—Å–ª–æ –¥–ª—è –∞–Ω–∞–ª–∏–∑–∞ –∏–ª–∏ '–¥–∏–∞–ø–∞–∑–æ–Ω' –¥–ª—è –º–∞—Å—Å–æ–≤–æ–π –≤–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏–∏")
    choice = input("–í–∞—à –≤—ã–±–æ—Ä (—á–∏—Å–ª–æ / '–¥–∏–∞–ø–∞–∑–æ–Ω'): ")

    if choice.isdigit():
        analyze_and_plot(int(choice))
    elif choice.lower() == '–¥–∏–∞–ø–∞–∑–æ–Ω':
        max_n = int(input("–î–æ –∫–∞–∫–æ–≥–æ n —Å—Ç—Ä–æ–∏—Ç—å –≥—Ä–∞—Ñ–∏–∫? (–Ω–∞–ø—Ä. 1000): "))
        batch_stability_plot(max_n=max_n)
    else:
        print("‚õî –ù–µ–∫–æ—Ä—Ä–µ–∫—Ç–Ω—ã–π –≤–≤–æ–¥. –ü–æ–ø—Ä–æ–±—É–π—Ç–µ —Å–Ω–æ–≤–∞.")

if __name__ == "__main__":
    main()
```
