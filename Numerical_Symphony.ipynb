```python
!pip install matplotlib rich

import matplotlib.pyplot as plt
from rich.console import Console
from rich.table import Table

console = Console()

def рассчитать_суммы(n1: int, n2: int):
    сумма_квадратов = sum(i**2 for i in range(n1, n2 + 1))
    сумма_кубов = sum(i**3 for i in range(n1, n2 + 1))
    разность = сумма_кубов - сумма_квадратов
    return сумма_квадратов, сумма_кубов, разность

def вывести_таблицу(n1, n2, сумма_квадратов, сумма_кубов, разность):
    table = Table(title="Анализ суммы квадратов и кубов", title_style="bold cyan")
    table.add_column("Диапазон", justify="center", style="green")
    table.add_column("Сумма квадратов", justify="center", style="yellow")
    table.add_column("Сумма кубов", justify="center", style="blue")
    table.add_column("Разность", justify="center", style="red")

    table.add_row(f"{n1} – {n2}", str(сумма_квадратов), str(сумма_кубов), str(разность))
    console.print(table)

def построить_график(n1, n2):
    числа = list(range(n1, n2 + 1))
    квадраты = [i**2 for i in числа]
    кубы = [i**3 for i in числа]

    plt.figure(figsize=(10, 6))
    plt.plot(числа, квадраты, label='Квадраты', color='orange', linewidth=2)
    plt.plot(числа, кубы, label='Кубы', color='purple', linewidth=2)
    plt.title('График квадратов и кубов чисел', fontsize=14)
    plt.xlabel('Число')
    plt.ylabel('Значение')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

def основная_программа():
    console.print("\n[bold green]Добро пожаловать в программу анализа суммы квадратов и кубов![/bold green]\n")

    try:
        n1 = int(console.input("Введите начальное число диапазона: "))
        n2 = int(console.input("Введите конечное число диапазона: "))

        if n1 > n2:
            console.print("[red]Ошибка: начальное число больше конечного![/red]")
            return

        сумма_квадратов, сумма_кубов, разность = рассчитать_суммы(n1, n2)
        вывести_таблицу(n1, n2, сумма_квадратов, сумма_кубов, разность)
        построить_график(n1, n2)

    except ValueError:
        console.print("[red]Ошибка: введите корректное целое число![/red]")

if __name__ == "__main__":
    основная_программа()
```
