✅ **Что делает программа:**
1. Строит модуль дзета-функции Римана на критической прямой;

2. Визуализирует "Имедин фильтр" для предсказания нетривиальных нулей;

3. Отображает это всё в красивом интерактивном графике.



```python
!pip install matplotlib numpy mpmath

import matplotlib.pyplot as plt
import numpy as np
from mpmath import zetazero, zeta
import matplotlib.ticker as ticker

# 📊 Настройка диапазона значений
t_values = np.linspace(0.1, 50, 1000)

# 📉 Вычисляем модуль дзета-функции Римана на критической прямой (Re = 0.5)
zeta_mod = [abs(zeta(0.5 + 1j * t)) for t in t_values]

# 🔶 Имедов фильтр (функция с затухающими осцилляциями)
omega = 0.55
alpha = 0.015
I_t = np.sin(omega * t_values) * np.exp(-alpha * t_values)

# 🎯 Известные нули дзета-функции (первые 5)
known_zeros = [zetazero(n).imag for n in range(1, 6)]

# 🎨 Создание графика
fig, ax = plt.subplots(figsize=(14, 7))
ax.plot(t_values, zeta_mod, label=r"$|\zeta(1/2 + it)|$ (модуль дзета)", color="navy", linewidth=2)
ax.plot(t_values, I_t, label=r"$I(t)$ (фильтр Имеды)", color="orange", linestyle="--", linewidth=2)

# 🚩 Отметим известные нули
for zero in known_zeros:
    ax.axvline(x=zero, color="red", linestyle=":", linewidth=1)

# 🏷️ Оформление графика
ax.set_title("🎯 Аналитическая визуализация: $|\zeta(1/2 + it)|$ и Имедин фильтр", fontsize=18, pad=20)
ax.set_xlabel("t (частота сигнала)", fontsize=14)
ax.set_ylabel("Значение функции", fontsize=14)
ax.legend(fontsize=12)
ax.grid(True, linestyle="--", alpha=0.6)

# 📏 Интервалы
ax.xaxis.set_major_locator(ticker.MultipleLocator(5))
ax.yaxis.set_major_locator(ticker.MultipleLocator(1))
ax.set_xlim(0, 50)

plt.tight_layout()
plt.show()
```

🧠 **Что делает программа**

1. **Вычисление модуля дзета-функции Римана:**
   
   *   Используется формула zeta(0.5 + i·t), где t — это переменная по оси X.
   *  Эта функция важна в теории чисел и криптографии.
2. **Имедов фильтр:**
   *   Используется специальная модель I(t) = sin(ωt) * exp(-αt) — функция, которая имитирует поведение нулей дзета-функции.
   *   Эта функция действует как "резонансный индикатор", помогающий находить особые точки (нули).
3. **Сравнение на графике**
   *   Строится график двух функций:
      *   **Синяя кривая** — реальная дзета-функция.
      *   **Оранжевая пунктирная** — Имедин фильтр.
   *   **Красные вертикальные линии** — места, где находятся известные нетривиальные нули дзета-функции.
4. **Визуализация:**
   *   Результат — красивый, информативный график, который показывает, как резонансные сигналы могут совпадать с математическими особенностями (например, нулями функции).
